## Function to extract episode from dataframe 

```python
def extract_ep(txt):
    check=False
    extracted=''
    for i in txt:
        if i==')':
            return extracted[1:]
        if i=='(':
            check=True
        if check==True:
            extracted+=i
    
            
     

    
```

```python
df['episodes']=df['Title'].apply(extract_ep)
df['episodes']
```

```python
df['episodes']=df['episodes'].str.replace(' eps','')
```

```python
df['episodes']=df['episodes'].astype(int)
```

## Function to extract Month range of anime episodes/Air duration

```python
def extract_air(txt):
    check=False
    data=""
    for i in range(len(txt)):
        if txt[i]==')':
            for j in range(i+1,i+20):
                data+=txt[j]
            return data
```

```python
df['air-duration']=df['Title'].apply(extract_air)
df.head()
```

```python
# Split start and end dates
df[['start', 'end']] = df['air-duration'].str.split(' - ', expand=True)

# Convert to datetime
df['start'] = pd.to_datetime(df['start'], format='%b %Y')
df['end'] = pd.to_datetime(df['end'], format='%b %Y')

# Calculate duration in months (inclusive)
df['months'] = (
    (df['end'].dt.year - df['start'].dt.year) * 12
    + (df['end'].dt.month - df['start'].dt.month)
    + 1
)

df.head()
```
